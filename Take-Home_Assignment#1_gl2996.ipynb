### Problem 1: Count vowels in a word
Write a program that takes a word as an input and print the number of vowels in the word.

````python
word = input().strip()
vowels = set('aeiouAEIOU')
count = sum(1 for ch in word if ch in vowels)
print(count)
````

### Problem 2: Print animals in ALL CAPS
Iterate through the following list of animals and print each one in all caps.

```animals=['tiger', 'elephant', 'monkey', 'zebra', 'panther']```

````python
animals=['tiger', 'elephant', 'monkey', 'zebra', 'panther']
for a in animals:
    print(a.upper())
````

### Problem 3: 1..20 odd/even
Write a program that iterates from 1 to 20, printing each number and whether it's odd or even.

````python
for n in range(1, 21):
    if n % 2 == 0:
        print(n, 'even')
    else:
        print(n, 'odd')
````

### Problem 4: Palindrome check
Write a program to check if a string is a palindrome or not.

````python
s = input().strip()
if s == s[::-1]:
    print('Palindrome')
else:
    print('Not Palindrome')
````
